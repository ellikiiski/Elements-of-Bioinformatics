## 1.1. Helper functions for min-cost flows

We will use the NextworkX Python library (https://networkx.org) for representing graphs, and for solving min-cost flow problems on them. Since its NetworkSimplex min-cost implementation (https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.flow.network_simplex.html#networkx.algorithms.flow.network_simplex) has a slightly different (but equivalent) interface to min-cost flows, we start with the following function `min_cost_flow(G, s, t)`. 

```python
import networkx as nx

maxInt = 1 << 32
newNodeId = 0

def getNewNodeId():
    global newNodeId
    newNodeId = newNodeId + 1
    return str(newNodeId)

def min_cost_flow(G, s, t):
    flowCost = 0
    flowValue = 0
    flowDict = dict()
    flowNetwork = nx.DiGraph()
    edgeMap = {}
    
    for v in G.nodes():
        if v == s:
            flowNetwork.add_node(v, demand = -maxInt)
        if v == t:
                flowNetwork.add_node(v, demand = maxInt)
        if v != s and v != t:
            flowNetwork.add_node(v, demand = 0)
    
    flowNetwork.add_edge(s, t, weight = 0)

    for (x,y) in G.edges():
        z1 = getNewNodeId()
        z2 = getNewNodeId()
        edgeMap[(x,y)] = [x,z1]
        l = G[x][y]['l']
        u = G[x][y]['u']
        c = G[x][y]['c']
        flowNetwork.add_node(z1, demand = l)
        flowNetwork.add_node(z2, demand = -l)
        flowNetwork.add_edge(x, z1, weight = c, capacity = u)
        flowNetwork.add_edge(z1, z2, weight = 0, capacity = u)
        flowNetwork.add_edge(z2, y, weight = 0, capacity = u)

    flowCost, flowDict0 = nx.network_simplex(flowNetwork)

    for x in G.nodes():
        flowDict[x] = dict()

    for (x,y) in G.edges():
        flowDict[x][y] = flowDict0[edgeMap[(x,y)][0]][edgeMap[(x,y)][1]]

    return flowCost, flowDict
```

The function above accepts in input a `nx.DiGraph` graph `G`, a so-called *directed* graph. All graphs used in the lectures were also directed, only that we have called them simply *graphs*. It also takes two nodes, `s` (the source of the flow network) and `t` (the sink of the flow network). It assumes that the edges of the graph have functions `l` (lowerbound), `u` (upperbound), `c` (cost), defined for each of them. It returns the flow cost `flowCost` (i.e. a number), and a function `flowDict` (i.e. a dictionary) assigning to every edges of `G` its flow value, in a mininimum cost flow (more on this below).

Let's start with a very simple example, just to see how we can use the `min_cost_flow` function.

First, we construct a directed graph with three nodes and two edges. Notice how we assign the `l`, `u`, `c` values to the edges. The following code will construct this flowNetwork:

## 2.2 Matching assembled sequences with reference sequences

We now implement the second reduction from the video lectures. Remember that:
- Each $a_i$ node (assembled sequenced), has to be matched with **exactly** one reference sequence (some $b_j$ node), or be classified as novel (star node).
- Each $b_j$ node (reference sequence) can be matched with **at most** one assembled sequence.

The task is to find such a matching satisfying the above two conditions, of total minimum cost.

```python
fg = open("data/graph2.edgelist", "rb")
G3 = nx.read_edgelist(fg, create_using=nx.DiGraph)
fg.close()

graph_dot = graph2dot(G3, edgeLabels = 'cost')
graph_dot
```

However, we assume the assembled sequences are those without in-neighbors (i.e. in-degree zero), and the reference sequences are those without out-neighbors (i.e. out-degree zero). 

Also, we assume that the node 'star' is the novel sequences, allowed to match many assembled sequences. Note that we already assigned as 10 the cost of the edges in-coming to 'star'.

```python
assembledSeqs = []
referenceSeqs = []

for v in G3.nodes():
    if G3.in_degree(v) == 0:
        assembledSeqs.append(v)
    if G3.out_degree(v) == 0:
        referenceSeqs.append(v)

print(assembledSeqs)
print(referenceSeqs)
```

However, it would be much more useful to have a visual representation of these paths. Let's write another graphviz function, that prints both the graph (with default balck edges), and each path, with a different color.

```python
def graphpaths2dot(nxDigraph, paths, pathWeights=None):
    colors = ['red','blue','green','purple','brown']

    dot = Digraph()
    dot.graph_attr['rankdir'] = 'LR' # Display the graph in landscape mode
    dot.node_attr['shape'] = 'rectangle' # Rectangle nodes

    for (u,v) in nxDigraph.edges():
        dot.edge(str(u),str(v)) 
         
    for index, path in enumerate(paths):
        pathColor = colors[index % len(colors)]
        if pathWeights == None:
            dot.attr('edge', color=pathColor, penwidth='2.0')
        else:
            dot.attr('edge', label=str(pathWeights[index]), fontcolor=pathColor, color=pathColor, penwidth='2.0')
        
        
        # If a path has only one node, color that node
        if len(path) == 1:
            dot.node(str(path[0]), color=colors[index % len(colors)], penwidth='2.0')
            continue
            
        # Otherwise
        prev_node = None
        for node in path:
            if prev_node == None:
                prev_node = node
                continue
            dot.edge(str(prev_node), str(node))
            prev_node = node
     
    return dot

G4_paths_dot = graphpaths2dot(G4, true_transcripts)
G4_paths_dot
```

## 3. Decomposing a flow into paths

**For this section, please watch the videos for lecture 2.4 (especially videos 3 & 4)**

As you see and realize from the above, the flow values alone do not give the paths (and thus the transcripts). However, any flow in a directed acyclic graph can be decomposed into a collection of weighted paths. As we've seen in the video lectures, finding the *minimum* number of such weighted paths is an NP-hard problem (thus unlikely to be solved efficiently). Moreover, we studied a heuristic for it, namely the one always choosing an $s$-$t$ path of maximum bottleneck.

The interesting part about the minimum path cover reduction, is that *any* path decomposition of the resulting min-cost flow will correspond to a flow decomposition into a minimum number of paths, and each such path will have weight 1. However, since we will also use the maximum bottleneck heuristic as a model of RNA transcript assembly, we implement it here.

## 3.1 DP for finding a path of maximum bottleneck

In this section we implement the DP algorithm from video lectures, which, given a flow, it finds an $s$-$t$ path of maximum bottleneck.

<h3 style='background:yellow;padding:10px;color:black'><strong>TASK 7</strong></h3>

Explain the following code (both for computing the B values, and for recovering the path of maximum bottleneck).

```python
def maxBottleckPath(flowNetwork, flowDict):
    topological_order = list(nx.topological_sort(flowNetwork))
    B = dict()
    maxInNeighbor = dict()

    # Computing the B values with DP
    for v in topological_order:
        if v == 's':
            B[v] = maxInt
            continue
        B[v] = -maxInt
        for u in flowNetwork.predecessors(v):
            uBottleneck = min(B[u], flowDict[u][v])
            if uBottleneck > B[v]:
                B[v] = uBottleneck 
                maxInNeighbor[v] = u
           
    # If not flow exists in the network
    if B['t'] == 0:
        return None, None
    
    # Recovering the path of maximum bottleneck
    v = 't'
    path = [v]
    while v != 's':
        v = maxInNeighbor[v]
        path = [v] + path
        
    return B['t'], path
```

## 3.2. Decomposing a flow, one path at a time

In this section, we implement a function that iteratiely finds a path of maximum bottleneck, and removes it from the flow.

<h3 style='background:yellow;padding:10px;color:black'><strong>TASK 8</strong></h3>

Explain the following code (both for computing the B values, and for recovering the path of maximum bottleneck).

```python
import copy

def decompose(flowNetwork, flowDict):
    pathDecomposition = []
    pathWeights = []
    # Making a copy of flowDict, otherwise the changes we make in this function 
    # will carry over to the global variable flowDict
    tempFlow = copy.deepcopy(flowDict)

    bottleneck, path = maxBottleckPath(flowNetwork, tempFlow)
    while path != None:
        prev_node = path[0]
        for node in path[1:]:
            tempFlow[prev_node][node] -= bottleneck
            prev_node = node
        
        pathDecomposition.append(path)
        pathWeights.append(bottleneck)
        bottleneck, path = maxBottleckPath(flowNetwork, tempFlow)
        
    return pathDecomposition, pathWeights
```

```python
def convertPaths(paths):
    convertedPaths = []

    for path in paths:
        convertedPath = []
        # Your code here
        # BEGIN SOLUTION
        for v in path:
            if v == 's' or v == 't':
                continue
            if v[-1] == '0':
                convertedPath.append(v[:-1])
                continue
            if v[-1] == '1':
                continue
            convertedPath.append(v)

        # END SOLUTION
        convertedPaths.append(convertedPath)
    return convertedPaths

convertedPaths4 = convertPaths(decomposedPaths4)
print('We assembled %d paths' % len(convertedPaths4))
G4_decomposed_dot = graphpaths2dot(G4, convertedPaths4)
G4_decomposed_dot
```

At the lecture, we saud that from the input coverage values, we construct a flow network as follows.

- We say that a node is a **candidate transcript start** if it has more out-going flow than in-coming flow. Namely, if the sum of coverage values of out-going edge is strictly greater than the sum of coverage of in-coming edges.
- We say that a node is a **candidate transcript end** if it has more in-going flow than out-coming flow. Namely, if the sum of coverage values of in-going edge is strictly greater than the sum of coverage of out-coming edges.

With this classification of nodes, we do as follows:
- We add each node and each edge of the graph to the flow network, with flow value equaling its coverage.
- We add a global source `s` and a global sink `t`.
- For every node `v` that is a candidate transcript start, we add the edge `(s,v)` with flow value equal to the sum of the out-going flow *minus* the sum of the in-coming flow.
- For every node `v` that is a candidate transcript end, we add the edge `(v,t)` with flow value equal to the sum of the in-coming flow *minus* the sum of the out-going flow.

<h3 style='background:yellow;padding:10px;color:black'><strong>TASK 10</strong></h3>

Let's build this network, call it `fN5`.

**NOTE**: Here, you don't need to set `l`, `u` and `c` values for each edge of the network, since we will not solve a min-cost flow problem on it. Instead, we will also construct a dictionary `flowDict5` storing the flow value mentioned above, so that we can call our function `decompose(fN5, flowDict5)` implemented above.

```python
fN5 = nx.DiGraph()

# We implement two functions testing if a node v is a candidate transcript start or end
def isCandidateStart(G,v):
    in_flow = sum([G[u][v]['coverage'] for u in list(G.predecessors(v))])
    out_flow = sum([G[v][w]['coverage'] for w in list(G.successors(v))])
    return (out_flow > in_flow), out_flow - in_flow

def isCandidateEnd(G,v):
    in_flow = sum([G[u][v]['coverage'] for u in list(G.predecessors(v))])
    out_flow = sum([G[v][w]['coverage'] for w in list(G.successors(v))])
    return (out_flow < in_flow), in_flow - out_flow

# We introduce the global source and sink
fN5.add_node('s')
fN5.add_node('t')

# We initialize the flowDict
flowDict5 = dict()
flowDict5['s'] = dict()
for v in G4.nodes():
    flowDict5[v] = dict()

# Implement the rest of the flow network
# Fill in the missing code fragments

for (u,v) in G4.edges():
    fN5.add_edge(u, v)
    # Your code here
    # BEGIN SOLUTION
    flowDict5[u][v] = G4[u][v]['coverage']
    # END SOLUTION

for v in G4.nodes():
    fN5.add_node(v)
    # Your code here
    # BEGIN SOLUTION
    isStart, flow = isCandidateStart(G4,v)
    if isStart:
        fN5.add_edge('s', v)
        flowDict5['s'][v] = flow
    isEnd, flow = isCandidateEnd(G4,v)
    if isEnd:
        fN5.add_edge(v, 't')
        flowDict5[v]['t'] = flow
    # END SOLUTION

fN5_dot = graph2dot(fN5)
fN5_dot
```

Let us now decompose `fN5` and `flowDict5` that you have built above using the heuristic of always selecting the path of maximum bottleneck. 

```python
decomposedPaths5, pathWeights5 = decompose(fN5, flowDict5)
convertedPaths5 = convertPaths(decomposedPaths5)

G5_decomposed_dot = graphpaths2dot(G4, convertedPaths5, pathWeights5)
G5_decomposed_dot
```
